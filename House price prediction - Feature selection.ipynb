```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('X_train.csv')
```

```python
dataset.head()
```

```python
y_train=dataset[['SalePrice']]
```

```python
X_train=dataset.drop(['Id','SalePrice'], axis=1)
```

```python
#apply feature selection

feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
# True indicates the particular feature must be used. False indicates otherwise
feature_sel_model.get_support()
```

```python
# Number of total and selected features

selected_features=X_train.columns[(feature_sel_model.get_support())]

print('total features: {}'.format(X_train.shape[1]))
print('selected features: {}'.format(len(selected_features)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
selected_features
```

```python
X_train=X_train[selected_features]
```

```python
import xgboost
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test=pd.read_csv('X_test.csv')
X_test.head()
```

```python
classifier=xgboost.XGBRFRegressor()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
pred=pd.DataFrame(y_pred)
pred = pd.DataFrame(y_pred, columns = ['logPrice'])
pred['PredPrice']=np.exp(pred)

pred=pred.drop(['logPrice'], axis=1)
pred
```

```python
sub_df=pd.read_csv('sample_submission.csv')
datasets=pd.concat([sub_df['Id'], pred], axis=1)
datasets.columns=['Id', 'SalePrice']
datasets.to_csv('submission.csv', index=False)
```
